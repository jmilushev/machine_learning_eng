### Analyze the training dataset 

```python
driver_images = pd.read_csv('data/driver_imgs_list.csv')
```

```python
driver_images.head()
```

```python
sns.catplot(x="classname", kind="count", palette="ch:.25", 
            data=driver_images, col="subject", col_wrap=6, height=2.5, aspect=.8, sharey=True, sharex=False);
```

```python
sns.catplot(x="subject", kind="count", palette="ch:.25", 
            data=driver_images, height=3.5, aspect=5.2);
```
